```python
#This notebook looks at the 3 WOIs with the lowest peak zeta measurements

import phasma2 as phasma
import astropy.units as u
import matplotlib.pyplot as plt

tic = 234494090
period = 0.972626507 * u.day
period_dimless = 0.972626507
event_duration = 1 * u.hr
event_epoch = 1325.50495
sectors = ['1', '2']

object_of_interest = phasma.Tess(tic, period, event_duration, event_epoch, sectors)
phase, time, flux, flux_err = (object_of_interest.phase,
                               object_of_interest._raw_time,
                               object_of_interest._raw_flux,
                               object_of_interest._raw_flux_err)

# this plots the phase curve
plt.errorbar(time % period_dimless, flux, flux_err, fmt='o')
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(data={'modt':time%period_dimless, 'flux':flux, 'flux_err':flux_err})
outside = df[(df['modt'] < 0.145) | (df['modt'] > 0.21)]
inside = df[(df['modt'] > 0.145) & (df['modt'] < 0.21)]

a = 0.0
b = 0.0
for row in inside.itertuples():
    a += row[2] * (row[3]**(-2))
    b += row[3]**(-2)
weighted_mean = a/b
print(weighted_mean)
```

```python
plt.errorbar(time % period_dimless, flux, flux_err, fmt='o', color='gray', alpha=0.2)
plt.plot([0.145, 0.145], [0.75, 1.05])
plt.plot([0.21, 0.21], [0.75, 1.05])
plt.plot([min(time%period_dimless), max(time%period_dimless)], [weighted_mean, weighted_mean])
plt.show()
```

```python
plt.errorbar(time % period_dimless, flux, flux_err, fmt='o', color='gray', alpha=0.2)
plt.plot([0.145, 0.145], [0.75, 1.05])
plt.plot([0.21, 0.21], [0.75, 1.05])
plt.plot([min(time%period_dimless), max(time%period_dimless)], [weighted_mean, weighted_mean])
plt.xlim(0.1, 0.3)
plt.show()
```

```python
print("<errorbar inside dip> / <# of points inside dip> = " + str(weighted_mean/len(inside['flux'])))
```

```python
import numpy as np
print("SNR = " + str(np.mean(inside['flux'])/(weighted_mean/len(inside['flux']))))
```
