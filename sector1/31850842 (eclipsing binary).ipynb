```julia
#With NO detrendeing

df0 = loadFITS(0, fitsdir="./Sample fits files/", tic_id="31850842", nodetrend=true)
d0 = pointsify(df0)
output0=periodogram(d0, periods, parallel=true, datakw=true)
output0[:delt_chi2] = flatten(periods, output0[:chi2], tess=true)
null_output0 = scrambled_periodogram(df0, periods, tess=true)
null_output0[:delt_chi2] = flatten(periods, null_output0[:chi2], tess=true)
sigma0 = movingstd((null_output0[
    :kurtosis] .- 3) .* (null_output0[:delt_chi2]))
    output0[:zeta] = (output0[:kurtosis] .- 3) .* output0[:delt_chi2] ./ sigma0

period = periods[findall(a->a==maximum(output0[:zeta]), output0[:zeta])][1]
df0[:modt] = df0[:t] .% period
scatter(df0[:modt], df0[:F], alpha=0.4, s=0.1)
```
