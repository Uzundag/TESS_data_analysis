```julia
using CSV

#first, the plots for kw=1
data = CSV.read("./kw1/sector1.csv")
logzeta = Float32[]
for x in data[:Peak_Zeta]
    push!(logzeta, log(10, x))
end
data[:logzeta] = logzeta
print(data)
```

```julia
using PyPlot
figure(figsize=(12, 9))
scatter(data[:Guessed_Period], data[:logzeta], alpha=0.8, s=0.3)
#Periods vs Log zeta
```

```julia
figure(figsize=(12, 9))
scatter(log.(data[:Guessed_Period]), data[:logzeta], alpha=0.8, s=0.3)
#Log Periods vs Log zeta
```

```julia
sec1tois = String[]
TOIs = CSV.read("../exofop_tess_tois.csv")
for i = 1:length(TOIs[:,1])
    if (typeof(TOIs[i, 45]) == typeof("") && contains(TOIs[i, 45], "1"))
        push!(sec1tois, TOIs[i, 1][1:7])
    end
end

count=0
figure(figsize=(12, 9))
for i = 1:length(data[:,1])
    if (string(data[i, 1])[1:7] in sec1tois)
        count += 1
        scatter(data[i, 3]), data[i, 4], alpha=0.8, s=1.2, color= "red")
    else
        scatter(data[i, 3], data[i, 4], alpha=0.8, s=0.3, color= "blue")
    end
end
print(count)
show()
```
