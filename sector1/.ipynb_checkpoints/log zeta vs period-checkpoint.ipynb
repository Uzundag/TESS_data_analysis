```julia
using CSV


#I first made the plots for kw=1
data = CSV.read("./kw=1/sector1.csv")
print(data)
```

```julia
logzeta = Float32[]
for x in data[:Peak_Zeta]
    push!(logzeta, log(10, x))
end
data[:logzeta] = logzeta
print(data)
```

```julia
using PyPlot
scatter(data[:Guessed_Period], data[:logzeta], alpha=0.8, s=0.3)
```

```julia
figure(figsize=(12, 9))
scatter(log.(data[:Guessed_Period]), data[:logzeta], alpha=0.8, s=0.3)
```

```julia
#Here I wanted to mark out the TOIs to look for any patterns

sec1tois = String[]
TOIs = CSV.read("../exofop_tess_tois.csv")
for i = 1:length(TOIs[:,1])
    if (typeof(TOIs[i, 45]) == typeof("") && contains(TOIs[i, 45], "1"))
        push!(sec1tois, TOIs[i, 1])
    end
end

figure()
for i = 1:length(data[:,1])
    if (string(data[i, 1]) in sec1tois)
        scatter(data[i, 3], data[i, 4], alpha=0.8, s=0.3, color= "red")
    else
        scatter(data[i, 3], data[i, 4], alpha=0.8, s=0.3, color= "blue")
    end
end
show()
```

```julia
scatter(data[:Guessed_Period], data[:logzeta], alpha=0.8, s=0.3)    
# List to hold x values.
x_number_values = [0, 15]

# List to hold y values.
y_number_values = [-7,-7]

# Plot the number in the list and set the line thickness.
plot(x_number_values, y_number_values, linewidth=3, color="green")

show()
```

```julia
notable = Float64[]
for i in 1:size(data, 1)
    if (data[i, 4] > -7)
        push!(notable, data[i, 1])
    end
end
println(notable)
print(length(notable))
```

```julia
c=0
for tid in notable
    for file in readdir("./foldedlightcurves")
        if contains(file, string(Int64(tid))[1:7])
            println(file)
            c+=1
            cp("./foldedlightcurves/"*file, "./notable/"*file, remove_destination=true)
        end
    end
end
print(c)
```
